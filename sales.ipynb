```python
df = df.rename(columns={'Product Category' : 'Product_Category','Price per Unit' : 'Price_per_Unit','Total Amount':'Total_Amount'})
```

```python
sns.boxplot(df['Total_Amount'])
```

```python
model_df = df[['Total_Amount','Price_per_Unit','Quantity','Product_Category']]
model_df
```

```python
import statsmodels.formula.api as smf
model = smf.ols('Total_Amount~+Product_Category+Quantity+ Price_per_Unit', data=model_df, missing='drop').fit()
model.summary()
```

```python
pred2 = model.predict(model_df)
rmse = np.sqrt(np.mean((np.array(model_df['Total_Amount']-np.array(pred2)))**2))
rmse
```

```python
from statsmodels.graphics.regressionplots import influence_plot
influence_plot(model)
plt.figure(figsize=(25,15))
plt.show()
```
